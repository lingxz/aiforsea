```python
import os
import pandas as pd
import numpy as np
```

```python
labels = pd.read_csv("data/labels/part-00000-e9445087-aa0a-433b-a7f6-7f4c19d78ad6-c000.csv")
```

```python
len(labels)
```

```python
dfs = [pd.read_csv("data/features/" + f) for f in os.listdir("data/features")]
features = pd.concat(dfs, ignore_index=True)
```

```python
features.groupby("bookingID").agg({"second": np.max}).reset_index()
```

```python
features[['bookingID']].groupby("bookingID").agg("count")
```

```python
from tsfresh import extract_features
from tsfresh.feature_extraction import MinimalFCParameters
```

```python
extracted_features = extract_features(features, column_id="bookingID", column_sort="second", default_fc_parameters = MinimalFCParameters())
```

```python
from tsfresh import select_features
from tsfresh.utilities.dataframe_functions import impute

impute(extracted_features)
```

```python
extracted_features.head()
```

```python
features_filtered = select_features(extracted_features, labels)
```

```python
combined = extracted_features.merge(labels, left_index=True, right_on="bookingID", how="left")
```

```python
x = combined.drop(['bookingID', 'label'], axis=1)
y = combined.label.values
```

```python
features_filtered = select_features(x, y)
```
